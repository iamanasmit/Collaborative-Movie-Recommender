```python
import pandas as pd
import numpy as np
```

```python
movies=pd.read_csv('movies.csv')
ratings=pd.read_csv('ratings.csv')
```

```python
movies.sample(5)
```

```python
ratings.head(7)
```

```python
# Now we create a pivot table
# 1 2 3 4 5 user_id
# 1
# 2
# 3
# 4
# movies_id
```

```python
final_dataset=ratings.pivot(index='movieId', columns='userId', values='rating')
```

```python
final_dataset.fillna(0, inplace=True)
final_dataset.head(5)
```

```python
#Removing users who filled less than 10 reviews and removing movies less than 50 reviews
```

```python
no_users_voted=ratings.groupby('movieId')['rating'].agg('count')
no_movies_voted=ratings.groupby('userId')['rating'].agg('count')
```

```python
no_users_voted.head(10)
```

```python
no_movies_voted.head(10)
```

```python
final_dataset=final_dataset.loc[no_users_voted[no_users_voted>10].index, :]
final_dataset.size
```

```python
final_dataset=final_dataset.loc[:, no_movies_voted[no_movies_voted>50].index]
final_dataset.shape
```

```python
final_dataset.head()
```

```python
from scipy.sparse import csr_matrix
csr_data=csr_matrix(final_dataset.values)
```

```python
final_dataset.reset_index(inplace=True)
```

```python
print(csr_data)
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(metric='cosine', algorithm='brute', n_neighbors=20, n_jobs=1)
model.fit(csr_data)
```

```python
movies.head()
```

```python
def get_recommendation(movie: str):
    recommedations = movies[movies['title'].str.contains(movie)]
    if len(recommedations) == 0:
        return 'not found'
    else:
        movie_idx = recommedations.iloc[0]['movieId']
        try:
            final_dataset[final_dataset['movieId'] == movie_idx].index[0]
        except:
            return 'filtered out'
        movie_idx = final_dataset[final_dataset['movieId'] == movie_idx].index[0]
        
        distances, indices = model.kneighbors(csr_data[movie_idx], n_neighbors=11)
        
        indices = indices.flatten()
        recommedations = movies[movies['movieId'].isin(final_dataset.iloc[indices]['movieId'])]
        
        print(distances, indices)
        return recommedations
```

```python
get_recommendation('Watchmen')
```
